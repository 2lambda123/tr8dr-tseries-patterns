```python
import pandas as pd
import numpy as np

from mvlabeler import AmplitudeBasedLabeler
```

# Data

```python
df = pd.read_csv ("csv/XAUUSD.csv", parse_dates=["stamp"])
```

```python
df.head()
```

# Labeling (minamp = 10bps, Tinactive = 5mins)

The data is 30sec bars, so Tinactive is set to 10 samples to indicate max inactive time of 5 minutes

```python
labeler = AmplitudeBasedLabeler (minamp = 10, Tinactive = 10)
labels = labeler.label (df)
labeler.plot()
```

# Labeling (minamp = 20bps, Tinactive = 5mins)

```python
labeler = AmplitudeBasedLabeler (minamp = 20, Tinactive = 10)
labels = labeler.label (df)
labeler.plot()
```

# Allowing move to ride more noise

## Labeling (minamp = 20bps, Tinactive = 15mins)

Increasing Tinactive to 15mins and minamplitude to 20bps allows for a move to ride through a significant amount of noise

```python
labeler = AmplitudeBasedLabeler (minamp = 20, Tinactive = 90)
labels = labeler.label (df)
labeler.plot()
```

## Labeling (minamp = 10bps, Tinactive = 15mins)

```python
labeler = AmplitudeBasedLabeler (minamp = 10, Tinactive = 90)
labels = labeler.label (df)
labeler.plot()
```
